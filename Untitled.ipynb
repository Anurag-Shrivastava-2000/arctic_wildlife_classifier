```python
import os
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_images_from_path(path, label):
    images, labels = [], []
    
    for file in os.listdir(path):
        img = image.load_img(os.path.join(path, file), target_size=(224, 224,
3))
        images.append(image.img_to_array(img))
        labels.append((label))
    return images, labels
```

```python
def show_images(images):
    fig, axes = plt.subplots(1, 8, figsize=(20, 20),
subplot_kw={'xticks': [], 'yticks': []})
    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i] / 255)
        
x_train, y_train, x_test, y_test = [], [], [], []
```

```python
images, labels = load_images_from_path('Wildlife/train/arctic_fox', 0)
show_images(images)
x_train += images
y_train += labels
```

```python
images, labels = load_images_from_path('Wildlife/train/polar_bear', 1)
show_images(images)
x_train += images
y_train += labels
```

```python
images, labels = load_images_from_path('Wildlife/train/walrus', 2)
show_images(images)
x_train += images
y_train += labels
```

```python
images, labels = load_images_from_path('Wildlife/test/arctic_fox', 0)
show_images(images)
x_test += images
y_test += labels
```

```python
images, labels = load_images_from_path('Wildlife/test/polar_bear', 1)
show_images(images)
x_test += images
y_test += labels
```

```python
images, labels = load_images_from_path('Wildlife/test/walrus', 2)
show_images(images)
x_test += images
y_test += labels
```

```python
import numpy as np
x_train = np.array(x_train) / 255
x_test = np.array(x_test) / 255
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
x_test.shape
```

```python
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.models import Sequential

model = Sequential()

model.add(Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
model.summary(line_length=80)
```

```python
hist = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=20, batch_size=10)

```

```python

acc = hist.history['accuracy']
val_acc = hist.history['val_accuracy']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, '-', label='Training Accuracy')
plt.plot(epochs, val_acc, ':', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.plot()
```

```python
x_train = preprocess_input(np.array(x_train)) / 255
x_test = preprocess_input(np.array(x_test)) / 255
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
base_model = ResNet50V2(weights='imagenet', include_top=False)
x_train = base_model.predict(x_train)
x_test = base_model.predict(x_test)
```

```python
model = Sequential()
model.add(Flatten())
model.add(Dense(1024, activation='relu'))
model.add(Dense(3, activation='softmax'))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
hist = model.fit(x_train, y_train,
validation_data=(x_test, y_test),
batch_size=10, epochs=10)
```

```python

acc = hist.history['accuracy']
val_acc = hist.history['val_accuracy']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, '-', label='Training Accuracy')
plt.plot(epochs, val_acc, ':', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.plot()
```
